```julia
using JuMP, Ipopt
```

```julia
# unbounded problem
model = Model(solver=IpoptSolver())


@defVar(model, x_var >= 0)
@defVar(model, y_var >= 0)

@setNLObjective( model, Min, -x_var + -y_var)
@addConstraint(model, x_var + y_var <= 1.0)
@addNLConstraint(model, (x_var)^2 + (y_var)^2 >= 2.0)
solve(model)
```

```julia
# unbounded problem
model = Model(solver=IpoptSolver())


@defVar(model, x_var >= 0)
@defVar(model, y_var >= 0)

@setNLObjective( model, Min, -x_var + -y_var)
#@addConstraint(model, x_var + y_var <= 1)
@addNLConstraint(model, x_var^2 + x_var == y_var)
@addNLConstraint(model, (x_var)^2 + (y_var)^2 >= 2)
solve(model)


```

```julia
# optimal problem
model = Model(solver=IpoptSolver())


@defVar(model, x_var >= 0)
@defVar(model, y_var >= 0)

@setNLObjective( model, Min, -x_var + -y_var)
@addConstraint(model, x_var + y_var <= 1)
@addNLConstraint(model, x_var^2 + x_var == y_var)
#@addNLConstraint(model, (x_var)^2 + (y_var)^2 >= 2)
solve(model)
```
